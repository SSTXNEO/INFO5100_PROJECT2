```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
# coding=utf-8
```

```python
df=pd.read_csv("./1-100.csv")
df
```

```python
j=496
for i in df['title'][496:500]:
    link="http://www.omdbapi.com/?t="+i+"&apikey=9baf85cd"
    r=requests.get(link)
    r_dict=json.loads(r.text)
    if(r_dict['Response']=='False'):
        print(j)
        continue
    else:
        print(r_dict['Actors'],j)
    j=j+1
# actors
```

```python
actors=pd.read_csv("./actors.csv")
# actors['actor4'][443]
changed_actor4=[]
for i in actors['actor4']:
    if(type(i)==float):
        i=' '
        changed_actor4.append(i)
        continue
    else:
        pieces=i.split(' ')
        i=' '.join(pieces[:-1])
        changed_actor4.append(i)
actors['actor4']=changed_actor4
actors.to_csv("./actors.csv")
```

```python
actors=pd.read_csv("./actors.csv")
actors_rank=[]
for i in actors['rank']:
    i=i+1
    actors_rank.append(i)
actors['rank']=actors_rank
actors
movie_actor=pd.merge(df,actors,on='rank')
movie_actor
```

```python
movie_actor.to_csv("./movie_actor.csv")
```
